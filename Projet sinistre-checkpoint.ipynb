```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
data["MONT"].values.reshape(-1,1)
data["AGE"].values.reshape(-1,1)
data["AGE"]=scaler.fit_transform(data["AGE"].values.reshape(-1,1))
data["MONT"]=scaler.fit_transform(data["MONT"].values.reshape(-1,1))

data[["AGE","MONT"]].plot.box()

#on utilise le StandardScaler par e que les valeur de l'age et du montant sont differentes 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_correlation_map( df ):
    

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 20 , 20) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )

plot_correlation_map(data)

# j'ai utiliser la matrice de correlation pour trouver la relation entre l'age et le perm la date d'obtenir le permis  
#et la relation entre le montant du sinistre et le perm 
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
#extract x and y from our data
x=data["AGE"].values[:,np.newaxis]
y=data["MONT"].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) #splitting data with test size of 35%

model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our model’s performance


print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted)) 

# the model explains none of the variance of the output around its mean 
```

```python
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("Montant Sinistre ")
plt.xlabel("Age")
plt.plot(x,model.predict(x),color="k")
plt.show()
```
